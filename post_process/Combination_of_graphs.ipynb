```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# List of image files to load
image_files = ['/home_student/kappen/pinns_for_comp_mech/post_process/pred_displacement_X_Beam2D_mixed_P_nonlinear_u_-0.1_swish_plane_strain.png', 
               '/home_student/kappen/pinns_for_comp_mech/post_process/pred_displacement_X_Beam2D_mixed_P_nonlinear_u_-0.1_swish_plane_strain.png', 
               '/home_student/kappen/pinns_for_comp_mech/post_process/pred_displacement_X_Beam2D_mixed_P_nonlinear_u_-0.1_swish_plane_strain.png', 
               '/home_student/kappen/pinns_for_comp_mech/post_process/pred_displacement_X_Beam2D_mixed_P_nonlinear_u_-0.1_swish_plane_strain.png']

# Custom titles for each image
titles = ['Graph 1', 'Graph 2', 'Graph 3', 'Graph 4']  # Customize these titles as needed

# Create a figure with subplots (4 rows, 2 columns)
fig, axs = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))  # Adjust figsize as needed

# Loop over image files and display them in both columns
for i, ax in enumerate(axs.flat):  # Flatten the 2D axs array to loop over it
    img = mpimg.imread(image_files[i % len(image_files)])  # Read each image (loop over image list)
    ax.imshow(img)  # Display image
    ax.axis('off')  # Remove axis labels for cleaner display
    ax.set_title(titles[i % len(titles)])  # Set custom title for each image (loop over titles)

# Adjust spacing between images (reduce space between subplots)
plt.subplots_adjust(hspace=0.75)  # Adjust this value for more/less space between images

# Show the combined plot
plt.show()
```
